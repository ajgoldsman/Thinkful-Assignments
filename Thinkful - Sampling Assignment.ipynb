```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Generate 2 binomially distributed populations
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 

print("\nPopulation 1 Mean: ")
print(pop1.mean())

print("\nPopulation 2 Mean: ")
print(pop2.mean())

print("\nPopulation 1 Std: ")
print(pop1.std())

print("\nPopulation 2 Std: ")
print(pop2.std())
```

```python
#Pull a sample of 100 entries from each population
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

print("\nSample 1 Mean: ")
print(sample1.mean())

print("\nSample 2 Mean: ")
print(sample2.mean())

print("\nSample 1 Std: ")
print(sample1.std())

print("\nSample 2 Std: ")
print(sample2.std())
```

1. Increase the size of your samples from 100 to 1000, then calculate the means and standard deviations for your new samples and create histograms for each. Repeat this again, decreasing the size of your samples to 20. What values change, and what remain the same?

```python
#Pull a sample of 1000 entries from each population
sample1 = np.random.choice(pop1, 1000, replace=True)
sample2 = np.random.choice(pop2, 1000, replace=True)

print("\nSample 1 Mean: ")
print(sample1.mean())

print("\nSample 2 Mean: ")
print(sample2.mean())

print("\nSample 1 Std: ")
print(sample1.std())

print("\nSample 2 Std: ")
print(sample2.std())

plt.hist(sample1)
line1 = plt.axvline(x=sample1.mean(), color = 'r', linestyle = '--')

for i in range(1,3):
    line2 = plt.axvline(x=(sample1.mean()+i*sample1.std()), color = 'm', linestyle = ':')
    plt.axvline(x=(sample1.mean()-i*sample1.std()), color = 'm', linestyle = ':')

plt.title("Sample 1 Binomial Distribution (Size = 1000) - Histogram\n")
plt.legend((line1, line2), ("Mean", "Standard Deviations"), loc='best')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()


plt.hist(sample2)
line1 = plt.axvline(x=sample2.mean(), color = 'r', linestyle = '--')

for i in range(1,3):
    line2 = plt.axvline(x=(sample2.mean()+i*sample2.std()), color = 'm', linestyle = ':')
    plt.axvline(x=(sample2.mean()-i*sample2.std()), color = 'm', linestyle = ':')

plt.title("Sample 2 Binomial Distribution (Size = 1000) - Histogram\n")
plt.legend((line1, line2), ("Mean", "Standard Deviations"), loc='best')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#Pull a sample of 20 entries from each population
sample1 = np.random.choice(pop1, 20, replace=True)
sample2 = np.random.choice(pop2, 20, replace=True)

print("\nSample 1 Mean: ")
print(sample1.mean())

print("\nSample 2 Mean: ")
print(sample2.mean())

print("\nSample 1 Std: ")
print(sample1.std())

print("\nSample 2 Std: ")
print(sample2.std())

plt.hist(sample1)
line1 = plt.axvline(x=sample1.mean(), color = 'r', linestyle = '--')

for i in range(1,3):
    line2 = plt.axvline(x=(sample1.mean()+i*sample1.std()), color = 'm', linestyle = ':')
    plt.axvline(x=(sample1.mean()-i*sample1.std()), color = 'm', linestyle = ':')

plt.title("Sample 1 Binomial Distribution (Size = 20) - Histogram\n")
plt.legend((line1, line2), ("Mean", "Standard Deviations"), loc='best')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()


plt.hist(sample2)
line1 = plt.axvline(x=sample2.mean(), color = 'r', linestyle = '--')

for i in range(1,3):
    line2 = plt.axvline(x=(sample2.mean()+i*sample2.std()), color = 'm', linestyle = ':')
    plt.axvline(x=(sample2.mean()-i*sample2.std()), color = 'm', linestyle = ':')

plt.title("Sample 2 Binomial Distribution (Size = 20) - Histogram\n")
plt.legend((line1, line2), ("Mean", "Standard Deviations"), loc='best')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

In this case, reducing the sample size didn't affect the mean values very significantly, but decreased the standard deviations for both samples. 

2. Change the probability value (p in the [NumPy documentation](https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.binomial.html)) for pop1 to 0.3, then take new samples and compute the t-statistic and p-value. Then change the probability value p for group 1 to 0.4, and do it again. What changes, and why?

```python
#Generate 2 binomially distributed populations, w/ p1=0.3 and p2=0.5
pop1 = np.random.binomial(10, 0.3, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 

#Pull a sample of 100 entries from each population
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

#Calculate t-value and p-value
from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```

```python
#Generate 2 binomially distributed populations, w/ p1=0.4 and p2=0.5
pop1 = np.random.binomial(10, 0.4, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 

#Pull a sample of 100 entries from each population
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

#Calculate t-value and p-value
from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```

The t-value decreased by nearly half and the p-value increased by a factor of more than e^10 when the probability value was increased from 0.3 to 0.4. The significantly smaller t-value for the second case (p=0.4) indicates that the means of the two samples are much closer together than those of the first case (p=0.3).
