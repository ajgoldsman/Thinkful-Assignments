```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Seaborn has several built-in datasets
sns.get_dataset_names()
```

```python
#Load chosen dataset from seaborn package
titanic = sns.load_dataset("titanic")
print(titanic.head())
```

- Choose one variable and plot that variable four different ways.

```python
#See how age data is spread
#Boxplot:
ax = sns.boxplot(titanic['age'], orient = 'horizontal') 
ax.set(xlabel = '\nAge')
ax.set(title = "Plot #1a: Boxplot of Titanic Passengers' Age\n")
plt.show()

#Violinplot
ax = sns.violinplot(titanic['age'])
ax.set(xlabel = '\nAge')
ax.set(title = "Plot #1b: Violinplot of Titanic Passengers' Age\n")
plt.show()

#Histogram
ax = sns.distplot(titanic['age'].dropna(), kde = False, norm_hist=False)
ax.set(xlabel = '\nAge', ylabel = 'Number of Passengers\n')
ax.set(title = "Plot #1c: Histogram of Titanic Passengers' Age\n")
plt.show()

#Kernel Density Estimate (KDE) Plot
ax = sns.kdeplot(titanic['age'])
ax.set(xlabel = '\nAge', ylabel = 'Kernel Desnity Estimate (KDE)\n of Passengers at Each Age\n')
ax.set(title = "Plot #1d: KDE Plot of Titanic Passengers' Age\n")
plt.show()
```

 - Choose two continuous variables, and plot them three different ways.

```python
#Because the only two continuous variables in this dataset are age and fare, we will work with those

#Unsorted Lineplot
ax = sns.relplot(x = 'age', y = 'fare', kind='line', sort = False,data = titanic)
ax.set(xlabel = '\nAge (years)', ylabel = 'Fare ($)\n')
ax.set(title = "Plot #2a: Unsorted Lineplot for Titanic - Passengers' Age vs. Fare Paid\n")
plt.show()

#Sorted Lineplot
ax = sns.relplot(x = 'age', y = 'fare', kind='line',data = titanic)
ax.set(xlabel = '\nAge (years)', ylabel = 'Fare ($)\n')
ax.set(title = "Plot #2b: Sorted Lineplot for Titanic - Passengers' Age vs. Fare Paid\n")
plt.show()

#Scatter
ax = sns.lmplot(x = 'age', y = 'fare', data = titanic, fit_reg=True, scatter_kws={'alpha':0.4}) 
ax.set(xlabel = '\nAge (years)', ylabel = 'Fare ($)\n')
ax.set(title = "Plot #2c: Scatterplot for Titanic - Passengers' Age vs. Fare Paid\n")
plt.show()

```

- Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
#Chosen variables: Age vs. Class (Are older people wealthier?)

#Boxplot
ax = sns.boxplot(x='class',y='age', data=titanic)  
ax.set(xlabel = '\nClass', ylabel = 'Age\n')
ax.set(title = "Plot #3a: Boxplots of Titanic Passengers' Age by Class\n")
plt.show()

#Violinplot
ax = sns.violinplot(x='class',y='age', data=titanic)
ax.set(xlabel = '\nClass', ylabel = 'Age\n')
ax.set(title = "Plot #3b: Violinplots of Titanic Passengers' Age by Class\n")
plt.show()

#Boxenplot
ax = sns.catplot(x='class',y='age', data=titanic, kind = 'boxen')
ax.set(xlabel = '\nClass', ylabel = 'Age\n')
ax.set(title = "Plot #3c: Boxenplots of Titanic Passengers' Age by Class\n")
plt.show()

#Barplot
ax = sns.catplot(x='class',y='age', data=titanic, height=6, kind="bar")
ax.despine(left=True)
ax.set(xlabel = '\nClass', ylabel = 'Age\n')
plt.title("Plot #3d: Barplots of Titanic Passengers' Age by Class\n")
plt.show()

#Pointplot
ax = sns.catplot(x='class',y='age', data=titanic, height=6, kind="strip")
ax.despine(left=True)
ax.set(xlabel = '\nClass', ylabel = 'Age\n')
plt.title("Plot #3e: Pointplots of Titanic Passengers' Age by Class\n")
plt.show()

#Histograms
ax = sns.FacetGrid(titanic, col="class")
# Name the plot type and the variable to be plotted using the structure.
ax.map(plt.hist, "age", lw=0)
# Moving the plots apart to make room for our titles.
plt.subplots_adjust(top=0.7)
# Making a more informative axis name.
ax.set_axis_labels('Age (in years)', 'Number of Passengers')
plt.suptitle("Plot #3f: Histograms of Age by Class")
# Removing excess lines around the plot.
sns.despine(trim=True)
plt.show()
```

## 2. Find a dataset with at least four continuous variables and one categorical variable. Create one master plot that gives insight into the variables and their interrelationships, including:
 - Probability distributions
 - Bivariate relationships
 - Whether the distributions or the relationships vary across groups
 - Accompany your plot with a written description of what you see.

```python
#Load chosen dataset from seaborn package
mpg = sns.load_dataset("mpg")
print(mpg.head())

np.unique(mpg['origin'])  #Categorical countries of origin
```

```python
#Categorical being used: origin
#Continuous being used: mpg, weight, acceleration, horsepower

#Plots separated by origin, x = horsepower, y = acceleration, markersize = weight, color = mpg
ax = sns.FacetGrid(mpg, col="origin")

def facet_scatter(x, y, c, s, **kwargs):
    kwargs.pop("color")
    plt.scatter(x, y, c=c, s=s/20, **kwargs)
    plt.colorbar() #for multiple colourbars
    
ax = (ax.map(facet_scatter, 'horsepower', 'acceleration', 'mpg', 'weight', cmap='Blues'))
ax.set_axis_labels('Horsepower', 'Acceleration')
plt.subplots_adjust(top=0.8)
plt.suptitle("Vehicles Masterplot: Markers Sized by Weight, Colored by MPG")
plt.show()
```
