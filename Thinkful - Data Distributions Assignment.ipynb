```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Laplace Distribution

laplace = np.random.laplace (1, 0.5, 100)
plt.hist(laplace)

avg = np.mean(laplace)
std = np.std(laplace)

line1 = plt.axvline(x=avg, color = 'r', linestyle = '--')

for i in range(1,3):
    line2 = plt.axvline(x=(avg+i*std), color = 'm', linestyle = ':')
    plt.axvline(x=(avg-i*std), color = 'm', linestyle = ':')

plt.title("Laplace Distribution - Histogram\n")
plt.legend((line1, line2), ("Mean", "Standard Deviations"), loc='best')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Logistic Distribution

logistic = np.random.logistic (5, 1, 100)
plt.hist(logistic)

avg = np.mean(logistic)
std = np.std(logistic)

line1 = plt.axvline(x=avg, color = 'r', linestyle = '--')

for i in range(1,3):
    line2 = plt.axvline(x=(avg+i*std), color = 'm', linestyle = ':')
    plt.axvline(x=(avg-i*std), color = 'm', linestyle = ':')

plt.title("Logistic Distribution - Histogram\n")
plt.legend((line1, line2), ("Mean", "Standard Deviations"), loc='best')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Rayleigh Distribution

rayleigh = np.random.rayleigh (10, 100)
plt.hist(rayleigh)

avg = np.mean(rayleigh)
std = np.std(rayleigh)

line1 = plt.axvline(x=avg, color = 'r', linestyle = '--')

for i in range(1,3):
    line2 = plt.axvline(x=(avg+i*std), color = 'm', linestyle = ':')
    plt.axvline(x=(avg-i*std), color = 'm', linestyle = ':')

plt.title("Rayleigh Distribution - Histogram\n")
plt.legend((line1, line2), ("Mean", "Standard Deviations"), loc='best')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Exponential Distribution

exponential = np.random.exponential (1, 100)
plt.hist(exponential)

avg = np.mean(exponential)
std = np.std(exponential)

line1 = plt.axvline(x=avg, color = 'r', linestyle = '--')

for i in range(1,3):
    line2 = plt.axvline(x=(avg+i*std), color = 'm', linestyle = ':')
    plt.axvline(x=(avg-i*std), color = 'm', linestyle = ':')

plt.title("Exponential Distribution - Histogram\n")
plt.legend((line1, line2), ("Mean", "Standard Deviations"), loc='best')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Geometric Distribution

geometric = np.random.geometric (0.3, 100)
plt.hist(geometric)

avg = np.mean(geometric)
std = np.std(geometric)

line1 = plt.axvline(x=avg, color = 'r', linestyle = '--')

for i in range(1,3):
    line2 = plt.axvline(x=(avg+i*std), color = 'm', linestyle = ':')
    plt.axvline(x=(avg-i*std), color = 'm', linestyle = ':')

plt.title("Geometric Distribution - Histogram\n")
plt.legend((line1, line2), ("Mean", "Standard Deviations"), loc='best')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Beta Distribution

beta = np.random.beta (1, 10, 100)
plt.hist(beta)

avg = np.mean(beta)
std = np.std(beta)

line1 = plt.axvline(x=avg, color = 'r', linestyle = '--')

for i in range(1,3):
    line2 = plt.axvline(x=(avg+i*std), color = 'm', linestyle = ':')
    plt.axvline(x=(avg-i*std), color = 'm', linestyle = ':')

plt.title("Beta Distribution - Histogram\n")
plt.legend((line1, line2), ("Mean", "Standard Deviations"), loc='best')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

The Laplace, Logistic, and Rayleigh distributions have generally normal distributions, as shown by the histograms, while Exponential, Geometric, and Beta Distributions are not normally distributed, in that the number of repetitions decreases with increasing x value. For the normally distributed ones, the mean and standard deviation provide useful information in that they tend to pinpoint the most common value in the series and show that most values fall within two standard deviations on each side, respectively. The descriptive statistics don't provide useful information for the three decreasing distributions, although the histograms do show similarities between them.

# Additional:

```python
mean1 = 5
std1 = 0.5

mean2 = 10
std2 = 1

norm1 = np.random.normal(mean1, std1, 100) 
norm2 = np.random.normal(mean2, std2, 100) 

norm3 = norm1 + norm2

plt.hist(norm3)
avg = np.mean(norm3)
std = np.std(norm3)

line1 = plt.axvline(x=avg, color = 'r', linestyle = '--')

for i in range(1,3):
    line2 = plt.axvline(x=(avg+i*std), color = 'm', linestyle = ':')
    plt.axvline(x=(avg-i*std), color = 'm', linestyle = ':')

plt.title("Combined Normal Distributions - Histogram\n")
plt.legend((line1, line2), ("Mean", "Standard Deviations"), loc='best')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

print("Mean: " + str(avg))
print("Standard Deviation: " + str(std))
```

The mean of the combined distribution was around 15, which is the sum of the means of the two variables (10 and 5) from which it was created. However, its standard deviation isn't the sum of the two standard deviations (0.5 and 1), but rather, approximately equal to the higher of the two values (1). I'm not sure, but perhaps this is a general trend for the combinations of normally distributed variables: the mean is the sum of the means, and the standard deviation is the higher of the two standard deviations.
